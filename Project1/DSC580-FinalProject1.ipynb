#2.) Perform any data extraction selection steps.

I really wanted to bin our ages so I can have a more useful metric to group this data before I input it into a machine learning model. I will bin them in groups of 25 below. 

```python
# Define the bin edges
bin_edges = [0, 25, 50, 75, 100]  # Specify the age intervals you want

# Define the labels for the bins
bin_labels = ['0-25', '26-50', '51-75', '76-100']

# Use pd.cut() to create a new column 'Age_Group' with the specified bins and labels
df['Age_Group'] = pd.cut(df['Age'], bins=bin_edges, labels=bin_labels, right=False)

# Drop the original 'Age' column if you no longer need it
df = df.drop('Age', axis=1)
df.head()
```
